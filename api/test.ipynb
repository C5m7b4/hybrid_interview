```python
whisky_last_names = [
    "Glenfiddich", "Macallan", "Laphroaig", "Ardbeg", "Glenlivet", "Lagavulin", "Balvenie", "HighlandPark", "Dalmore",
    "Glenmorangie", "Talisker", "Oban", "GlenDronach", "Bruichladdich", "Springbank", "Bowmore", "Auchentoshan",
    "Yamazaki", "Hakushu", "Hibiki", "Nikka", "Kavalan", "Redbreast", "Jameson", "Tullamore", "Knappogue", "Bushmills",
    "Glen Scotia", "Edradour", "BenRiach"
]

# Common first names
first_names = ["John", "Jane", "Jim", "Jack", "Jill", "Alice", "Bob", "Charlie", "Daisy", "Eve", "Frank", "Grace",
               "Henry", "Ivy", "Jake", "Kelly", "Liam", "Mia", "Noah", "Olivia", "Paul", "Quinn", "Rachel", "Steve",
               "Tina", "Umar", "Vera", "Walter", "Xena", "Yann", "Zoe"]

# Gender options
genders = ["M", "F"]
```

```python
def random_date(start_date="2005-01-01", end_date="2025-01-01"):
    start_dt = datetime.strptime(start_date, "%Y-%m-%d")
    end_dt = datetime.strptime(end_date, "%Y-%m-%d")
    random_days = random.randint(0, (end_dt - start_dt).days)
    return (start_dt + timedelta(days=random_days)).strftime("%Y-%m-%d")
```

```python
peeps = [
    (i + 1, random.choice(first_names), random.choice(whisky_last_names),
     random.randint(0, 1), random_date(), random.choice(genders))
    for i in range(50)
]
```

```python
df_peeps = pd.DataFrame(peeps, columns=["id", "first_name", "last_name", "employed", "hire_date", "gender"])
```
